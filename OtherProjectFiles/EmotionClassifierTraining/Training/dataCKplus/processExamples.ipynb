```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import csv
from random import shuffle

#################################################################################
# GENERATE EXAMPLES from BALANCED data, skipping 7 UNRECOGNISED AUs
# AVERAGE or CONCATENATE
# set here:
WINDOW = 1    # between 1 and 6
strategy = 'concat'
#strategy = 'avg'

# for 8 classes
# inExamples = './examples/exAllBalanced.dat'
# outExamples = './examples/ex7AU' + strategy + 'W' + str(WINDOW) + '.dat'
# for 4 classes
inExamples = './examples/exAll4Classes.dat'
outExamples = './examples/ex7AU4' + strategy + 'W' + str(WINDOW) + '.dat'
newExamples = []
examples = []
ACTIVE_AUS = 7
TOTAL_AUS = 12
# active ones: [1,2,4,6,12,-,-,-,25,26,-,-] => [0, 1, 2, 3, 4, 8, 9]
activeAUs = [0, 1, 2, 3, 4, 8, 9]

with open(inExamples, 'rb') as ff:
    for row in csv.reader(ff, delimiter=' '):
        rowFloat = []
        rowFloat = [np.float64(elem) for elem in row if elem is not '']
        examples.append(rowFloat)

for ex in examples:

    # get #frames x 12 matrix
    data = np.asarray(ex[:-1]).reshape(-1, TOTAL_AUS)
    # get WINDOW x 7 matrix
    data = data[-WINDOW:, activeAUs]
    
    if strategy == 'concat':
        tmp = data.flatten().tolist()
    elif strategy == 'avg':
        tmp = np.mean(data, axis=0).tolist()
    else:
        print "invalid input setting!"
        break
    
    tmp.append(ex[-1])
    newExamples.append( tmp )

    
if strategy == 'concat':
    newExamples = np.asarray(newExamples).reshape(-1, (ACTIVE_AUS * WINDOW) + 1)
elif strategy == 'avg':
    newExamples = np.asarray(newExamples).reshape(-1, ACTIVE_AUS + 1)
else:
    print "invalid input setting!"

print "Matrix of examples with dimensions: (%d,%d)" %(np.shape(newExamples))
f = open(outExamples, 'w')
np.savetxt(outExamples, newExamples, delimiter = ' ', fmt='%f')
f.close()

#################################################################################
```

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import csv

#################################################################################
# REMOVE 4 CLASSES ['anger'/1,'contempt'/2,'fear'4,'sadness'6] FROM 8 balanced classes:
# AND RELABEL to [0->0, 1->_, 2->_, 3->1, 4->_, 5->2, 6->_, 7->3]
# this removal is reasonable according to confusion matrix for balanced 8 classes
# 410 - 116 = 294 examples left
# no need to run anymore
# this is for all 12 AUs
# new classes: ['neutral', 'disgust', 'happiness', 'surprise']

inExamples = './examples/exAllBalanced.dat'
outExamples = './examples/exAll4Classes1.dat'
examples = []
TOTAL_AUS = 12

with open(inExamples, 'rb') as ff:
    for row in csv.reader(ff, delimiter=' '):
        rowFloat = []
        rowFloat = [np.float64(elem) for elem in row if elem is not '']

        if(rowFloat[-1] == 0.0):
            examples.append(rowFloat)
        elif(rowFloat[-1] == 3.0):
            # relabel 3 to 1
            rowFloat[-1] = 1.0
            examples.append(rowFloat)
        elif(rowFloat[-1] == 5.0):
            # relabel 5 to 2
            rowFloat[-1] = 2.0
            examples.append(rowFloat)
        elif(rowFloat[-1] == 7.0):
            # relabel 7 to 3
            rowFloat[-1] = 3.0
            examples.append(rowFloat)

print "%d examples created" %(len(examples))
f = open(outExamples, 'w')
np.savetxt(outExamples, examples, delimiter = ' ', fmt='%s')
f.close()

#################################################################################
```
