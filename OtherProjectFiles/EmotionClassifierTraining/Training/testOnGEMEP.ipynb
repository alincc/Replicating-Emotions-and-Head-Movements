```python
#######################################################################################################
# Author: Jan Ondras
# Institution: University of Cambridge
# Project: Replicating Human Facial Emotions and Head Movements on a Robot Avatar (Part II Project)
# Duration: October 2016 - May 2017
####################################################################################################### 
# Test the final trained model on much more challenging GEMEP database => produce Confusion matrix
#######################################################################################################

import time
import numpy as np 
import csv
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt

inFeatures = './dataGEMEP/features/features_GEMEP.dat'
inLabels = './dataGEMEP/labels/labels.dat'
inWeightsFile = "./weights.npz"

labels = []
normalizeCM = True
TOTAL_AUS = 12
activeAUs = [0, 1, 2, 3, 4, 8, 9]
classesModel = ['neutral', 'disgust', 'happiness', 'surprise'] # 0,1,2,3
classesGEMEP = ['anger', 'joy', 'fear', 'sadness', 'relief'] # 0,1,2,3,4
cm = np.zeros((len(classesGEMEP), len(classesModel))) # confusion matrix

# Load weights
weights = np.load(inWeightsFile)
z = np.zeros(len(weights['b1']))
strategy = weights['strategy']
WINDOW = weights['WINDOW']

# Load GEMEP data
# Compare predicted and true labels 
# Predicted is obtained by sliding window over whole video and then majority vote
with open(inLabels, 'rb') as fl:
	for row in csv.reader(fl):
		labels.append(classesGEMEP.index(row[0]))
	fl.close()

with open(inFeatures, 'rb') as ff:
    for i, row in enumerate(csv.reader(ff, delimiter='\t')):
        rowFloat = []
        rowFloat = [np.float64(elem) for elem in row if elem is not '']
        data = np.asarray(rowFloat[2:]).reshape(-1, TOTAL_AUS)
        data = data[:, activeAUs] #    (#frames in video x 7) matrix
        
        if len(data) < WINDOW:
            print "This example is too short (%d) for given window size (%d) (skipping)" %(len(data), WINDOW)
        else:
            votes = np.zeros(len(classesModel))
            # Sliding window over whole video
            for j in range(0, len(data) - WINDOW + 1):
                
                if strategy == 'avg':
                    X = np.mean(data[j:(j + WINDOW)], axis=0)
                else: # 'concat'
                    X = data[j:(j + WINDOW)].flatten()

                # Prediction
                f = np.dot(np.maximum(z, np.dot(X, weights['w1']) + weights['b1']), weights['w2']) + weights['b2'] # relu activation
                s = np.exp(f) / np.sum(np.exp(f), axis=0) # Softmax
                y = np.argmax(s) # predicted label
                votes[y] += 1 # record vote
                
            # Majority voting
            y_pred = np.argmax(votes)
        
            y_true = labels[i]
            cm[y_true, y_pred] += 1

    ff.close()

# Show confusion matrix
plt.figure()
if normalizeCM:
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    print("Normalized confusion matrix")
plt.title('Confusion matrix, testing on GEMEP database')
print cm

plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.colorbar()
tick_marksX = np.arange(len(classesModel))
tick_marksY = np.arange(len(classesGEMEP))
plt.xticks(tick_marksX, classesModel, rotation=45)
plt.yticks(tick_marksY, classesGEMEP)

thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    if normalizeCM:
        plt.text(j, i, '%.3f' % (cm[i, j]), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")	# fix text color
    else:
        plt.text(j, i, int(cm[i, j]), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")	# fix text color
plt.tight_layout()
plt.ylabel('True label from GEMEP')
plt.xlabel('Predicted label by a model')
plt.show()

weights.close()
```
