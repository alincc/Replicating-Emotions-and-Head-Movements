```python
#######################################################################################################
# Author: Jan Ondras
# Institution: University of Cambridge
# Project: Replicating Human Facial Emotions and Head Movements on a Robot Avatar (Part II Project)
# Duration: October 2016 - May 2017
####################################################################################################### 
# Estimation of measurement noise r = Variance
# For 5 different positions: (0,0), (-30,+30), (+30,+30), (+30,-30), (-30,-30) 
# and 3 different subjects (first 3 from DB)
# 1000 measurements each using HeadPoseImageDatabase by Gourier

# To generate measurements from live recording: 
# run as ./DemoTracker LIVE ./generatedData/estimateR_01_02.dat (subject 1, angle set 2)
# frame rate cca 47 FPS (irrelevant here)
#######################################################################################################



import time
import numpy as np
from matplotlib import pyplot as plt

# log variances of each measurement 
Y_vars = []
P_vars = []
N = 1000 # take first N measurements
for i in range(1, 4): # subjects
    for j in range(1, 6): # angles set
        path = format(i, '02d') + "_" + format(j, '02d')
        data = np.genfromtxt("./../HPDetector/generatedData/estimateR_" + path + ".dat", usecols = (0, 1))
        Y_vars.append(np.std(data[:(N+1),0])**2)
        P_vars.append(np.std(data[:(N+1),1])**2)

# just to see
plt.figure()
plt.plot([1,2,3,4,5]*3, Y_vars, 'o')
plt.plot([1,2,3,4,5]*3, P_vars, 'o')
plt.show()

print "Measurement noise for yaw is: "
print np.mean(Y_vars)
print np.std(Y_vars)
# rY = 8e-5 +/- 5e-5

print "Measurement noise for pitch is: "
print np.mean(P_vars)
print np.std(P_vars)
# rP = 6e-5 +/- 2e-5


# # For one measurement NOT NEEDED
# data = np.genfromtxt("./../HPDetector/generatedData/estimateR.dat", usecols = (0, 1))
# print np.mean(data[:,0])
# print np.mean(data[:,1])
# print np.std(data[:,0])
# print np.std(data[:,1])
# print np.std(data[:,0])*100 / np.mean(data[:,0])
# print np.std(data[:,1])*100 / np.mean(data[:,1])

# print "Variance yaw is: "
# print np.std(data[:,0]) ** 2

# print "Variance pitch is: "
# print np.std(data[:,1]) ** 2
```

```python
#######################################################################################################
# For finding optimal q (process noise) given r (already estimated)
# This minimises sum of squared errors over all datapoints from UPNA database
# Uses unconstrained KF
#######################################################################################################

import time
import numpy as np
from matplotlib import pyplot as plt
from KFClass import KFOnline
from scipy.optimize import minimize_scalar

rY = 8e-5
rP = 6e-5
dt = 1.0/30 # 30 FPS - given by UPNA

# No constraints
# x_yaw_max = np.array([2.0857, 8.26797])
# x_yaw_min = np.array([-2.0857, -8.26797])
# x_pitch_max = np.array([0.200015, 7.19407])
# x_pitch_min = np.array([-0.330041, -7.19407])

# Returns sum of squared errors (SSE) over all datapoints from UPNA database: sum((true - filtered)^2)
def getSSE(q, r, angleType):

    SSE = 0. # sum of squared errors
    
    # Compare on 10x12=120 videos:
    for i in range(1, 11): # i in range(1, 9):
        for j in range(1, 13):
            path = "user_" + format(i, '02d') + "_video_" + format(j, '02d')
            
            if angleType == 'yaw':
                dataTracked = np.genfromtxt("./UPNAtrackedZeroed/" + path + ".dat", usecols = 0)
                dataTrue = np.genfromtxt("../../Databases/Head_Pose_Database_UPNA/labelsZero/" + path + "_groundtruth3D_zeroed.txt", usecols = 4)
            
            elif angleType == 'pitch':
                dataTracked = np.genfromtxt("./UPNAtrackedZeroed/" + path + ".dat", usecols = 1)
                dataTrue = np.genfromtxt("../../Databases/Head_Pose_Database_UPNA/labelsZero/" + path + "_groundtruth3D_zeroed.txt", usecols = 5)
            else:
                print "Wrong argument 'angleType'!"
            
            # Filter the tracked
            #mykf = KFOnline(dt=dt, q=q, r=r, init_P_post=np.matrix('1 0 0; 0 1 0; 0 0 1'), init_x_est_post=np.matrix('0; 0; 0'), x_min=x_pitch_min, x_max=x_pitch_max)
            mykf = KFOnline(dt=dt, q=q, r=r, init_P_post=np.matrix('1 0 0; 0 1 0; 0 0 1'), init_x_est_post=np.matrix('0; 0; 0'))
            mykf.updateAll(dataTracked)
            dataFiltered = mykf.getEstPositionAsArray()
            
            # Convert ground truth values to RADIANS:
            dataTrue = dataTrue * np.pi/180
            
            # for 9 videos we have less than 300 frames ! => exclude these
            # This was caused probably by detector - not taking measurements for some frames (face was not detected)
            if len(dataTracked) == len(dataTrue):
                SSE += np.sum((dataFiltered - dataTrue)**2)
    return SSE # in radians^2

###################################################################################################
# Minimise SSE - over whole DB => for real-time
# i in range(1, 11)

res = minimize_scalar(getSSE, args=(rY, 'yaw'))
print res
# RESULTS YAW
#      fun: 285.25307044394492
#     nfev: 21
#      nit: 20
#  success: True
#        x: 1.7864724004181776

res = minimize_scalar(getSSE, args=(rP, 'pitch'))
print res
# RESULTS PITCH
#      fun: 54.746179266202233
#     nfev: 16
#      nit: 15
#  success: True
#        x: 0.63319015805369705

####################################################################################################
# EVALUATION
# ONLY FOR first 8/10 subjects from DB (this is for evaluation; the above is for real-time)
# i in range(1, 9)

# RESULTS YAW
#      fun: 219.90488565810904
#     nfev: 21
#      nit: 20
#  success: True
#        x: 1.4001736652373062

# RESULTS PITCH
#      fun: 28.163779726358094
#     nfev: 21
#      nit: 20
#  success: True
#        x: 0.30687422619976734


####################################################################################################
# Also tried with constrained KF, and the difference is neglible
# RESULTS YAW
#      fun: 285.253067445434
#     nfev: 22
#      nit: 21
#  success: True
#        x: 1.7864535560675179
# RESULTS PITCH
#      fun: 55.91678871776503
#     nfev: 39
#      nit: 38
#  success: True
#        x: 0.23990836688022299
```
