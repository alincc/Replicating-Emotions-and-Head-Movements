```python
#######################################################################################################
# Author: Jan Ondras
# Institution: University of Cambridge
# Project: Replicating Human Facial Emotions and Head Movements on a Robot Avatar (Part II Project)
# Duration: October 2016 - May 2017
####################################################################################################### 
# Compare filtered estimates from Kalman filter with truth data from DB UPNA
# Evaluation method: plot errors in terms of angle, 
# then quantized into bins and plotted mean with error-bars 
# and finally total mean and std over all angles

# UPNA: 30 FPS
#######################################################################################################

import time
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from KFClass import KFOnline

CREATE_ZEROED = False # run only once ...
N = 120
FPS = 30
qY = 1.40 #1.79
qP = 0.31 #0.63
rY = 8e-5
rP = 6e-5

# angles in RADIANS, from DB in DEGREES
# load data: measured yaw, pitch

# Create zeroed data and save, run only once ...
if CREATE_ZEROED:
	print "Created zeroed data"
	for i in range(1, 11):
		for j in range(1, 13):
			path = "user_" + format(i, '02d') + "_video_" + format(j, '02d')
			data = np.genfromtxt("./UPNAtracked/" + path + ".dat", usecols = (0, 1))

			# Substract first measurement from all the other ones => can compare with zeroed ground truth
			data = data - data[0]
			# Save as zeroed
			np.savetxt("./UPNAtrackedZeroed/" + path + ".dat", data, delimiter = " ")


print "111 videos with 300 frames each:"

yawX = []
yawY = []
pitchX = []
pitchY = []

# Compare 10x12=120 videos:
# Generate data yawX, yawY, pitchX, pitchY for later comparison
for i in range(9, 11): # i in range(1, 11):
	for j in range(1, 13):
		path = "user_" + format(i, '02d') + "_video_" + format(j, '02d')
		dataTracked = np.genfromtxt("./UPNAtrackedZeroed/" + path + ".dat", usecols = (0, 1))
		dataTrue = np.genfromtxt("../../Databases/Head_Pose_Database_UPNA/labelsZero/" + path + "_groundtruth3D_zeroed.txt", usecols = (4, 5))
		# Filter here; 
		mykf_Y = KFOnline(dt=1.0/FPS, q=qY, r=rY, init_P_post=np.matrix('1 0 0; 0 1 0; 0 0 1'), init_x_est_post=np.matrix('0; 0; 0'))
		mykf_P = KFOnline(dt=1.0/FPS, q=qP, r=rP, init_P_post=np.matrix('1 0 0; 0 1 0; 0 0 1'), init_x_est_post=np.matrix('0; 0; 0'))
		mykf_Y.updateAll(dataTracked[:,0])	# yaw
		mykf_P.updateAll(dataTracked[:,1])	# pitch
		dataFiltered = np.column_stack((mykf_Y.getEstPositionAsArray(), mykf_P.getEstPositionAsArray()))

		# Convert ground truth values to RADIANS:
		dataTrue = dataTrue * np.pi/180

		# for 9 videos we have less than 300 frames ! => exclude these
		# This was caused probably by detector - not taking measurements for some frames (face was not detected)
		if len(dataTracked) != 300:
			print "Shorter: user: %d video: %d tracked: %d true: %d" %(i, j, len(dataTracked), len(dataTrue))

		else:
			
			diff0 = dataTracked - dataTrue
			diff = dataFiltered - dataTrue
            
			yawX.extend(dataTrue[:,0])
			yawY.extend(diff[:,0])
			pitchX.extend(dataTrue[:,1])
			pitchY.extend(diff[:,1])

# 			if i == 1 and j == 7:

# 				plt.subplot(2, 1, 1)
# 				plt.plot(dataTrue[:,0], diff[:,0], label='error(filtered-truth) vs. true angle')
# 				plt.title('Yaw angle')
# 				plt.legend()

# 				# pitch
# 				plt.subplot(2, 1, 2)
# 				plt.stem(dataTrue[:,1], diff[:,1], label='a')
# 				plt.title('Pitch angle')
# 				plt.legend()

# 				plt.show()
```

```python
###########################################################################################
# Plot obtained data; quantized into bins
# set #bins_edges to 25&19 to make equal step size 0.057 rad

# YAW
bins_edges = np.linspace(np.min(yawX)-0.000001, np.max(yawX)+0.000001, 25) #small offset to handle roundings
step = bins_edges[1]-bins_edges[0]
xlimY = [bins_edges[0], bins_edges[-1]]
bins_centers = [np.mean([bins_edges[i], bins_edges[i + 1]]) for i in range(0, len(bins_edges) - 1)]
digitized = np.digitize(yawX, bins_edges)
bin_means = []
bin_stds = []
# iterate over bins (i) and then over elements in that bin
for i in range(1, len(bins_edges)):
    bin_i = []
    for j, element in enumerate(digitized):
        if element == i:
            bin_i.append(np.absolute(yawY[j]))    # take absolute errors
    bin_means.append(np.mean(bin_i))
    bin_stds.append(np.std(bin_i))
ylimY = np.min(np.array(bin_means)-np.array(bin_stds))-0.001
plt.figure()
plt.subplot(2, 1, 1)
plt.bar(bins_edges[:-1], bin_means, width=step, color='b', alpha=0.08, label='Step: %.3f rad' %step)
plt.errorbar(bins_centers, bin_means, yerr=bin_stds, fmt='go', 
             label='Total mean: %.3f +/- %.3f rad' %(np.mean(np.absolute(yawY)), np.std(np.absolute(yawY))))
plt.title('Yaw angle absolute errors')
plt.xlabel('yaw angle (rad)')
plt.ylabel('error |filtered - truth| (rad)')
plt.xlim(xlimY)
plt.ylim(ylimY)
plt.legend(loc='best')
#plt.show()


# PITCH
bins_edges = np.linspace(np.min(pitchX)-0.000001, np.max(pitchX)+0.000001, 19)
step = bins_edges[1]-bins_edges[0]
bins_centers = [np.mean([bins_edges[i], bins_edges[i + 1]]) for i in range(0, len(bins_edges) - 1)]
digitized = np.digitize(pitchX, bins_edges)
bin_means = []
bin_stds = []
# iterate over bins (i) and then over elements in that bin
for i in range(1, len(bins_edges)):
    bin_i = []
    for j, element in enumerate(digitized):
        if element == i:
            bin_i.append(np.absolute(pitchY[j]))    # take absolute errors
    bin_means.append(np.mean(bin_i))
    bin_stds.append(np.std(bin_i))
#plt.figure()
plt.subplot(2, 1, 2)
plt.bar(bins_edges[:-1], bin_means, width=step, color='b', alpha=0.08, label='Step: %.3f rad' %step)
plt.errorbar(bins_centers, bin_means, yerr=bin_stds, fmt='go', 
             label='Total mean: %.3f +/- %.3f rad' %(np.mean(np.absolute(pitchY)), np.std(np.absolute(pitchY))))
plt.title('Pitch angle absolute errors')
plt.xlabel('pitch angle (rad)')
plt.ylabel('error |filtered - truth| (rad)')
plt.xlim(xlimY)
plt.ylim(ylimY)
plt.legend(loc='best')
plt.show()



########################################################################################
# Plot all datapoints of errors (not absolute) - no longer needed
# plt.figure()

# # yaw - absolute
# plt.subplot(2, 2, 1)
# plt.plot(yawX, yawY, 'rx')
# plt.title('Yaw angle errors')
# plt.xlabel('yaw angle (rad)')
# plt.ylabel('error (filtered - truth) (rad)')
# plt.grid()

# # yaw - relative
# # plt.subplot(2, 2, 2)
# # plt.plot(yawX, np.array(yawY, dtype='float')/np.array(yawX), 'rx')
# # plt.title('Yaw angle relative errors')
# # plt.xlabel('yaw angle (rad)')
# # plt.ylabel('relative error')
# # plt.grid()

# # pitch
# plt.subplot(2, 2, 3)
# plt.plot(pitchX, pitchY, 'rx')
# plt.title('Pitch angle errors')
# plt.xlabel('pitch angle (rad)')
# plt.ylabel('error (filtered - truth) (rad)')
# plt.grid()

# # yaw - relative
# # plt.subplot(2, 2, 4)
# # plt.plot(pitchX, np.array(pitchY, dtype='float')/np.array(pitchX), 'rx')
# # plt.title('Pitch angle relative errors')
# # plt.xlabel('pitch angle (rad)')
# # plt.ylabel('relative error')
# # plt.grid()

# plt.show()
```
